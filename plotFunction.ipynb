```python
# Import
import matplotlib.pyplot as plt
```

```python
# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales_2023 = [1200, 1500, 1800, 1600, 2100, 2000, 2200, 2500, 2400, 2300, 1900, 1700]
sales_2024 = [1300, 1600, 1700, 1650, 2200, 2100,2300, 2550, 2450, 2350, 2000, 1800]
```

```python
plt.plot(months, sales_2023, color = 'purple', linestyle = '--', linewidth = 3, marker = 'o', label = "Sales 2023")
plt.xlabel("Months")
plt.ylabel("Sales in year 2023")
plt.title("Monthly sales data in FY 2023")
plt.legend()
plt.xlim(0,11)
plt.ylim(0,2600)
```

```python
plt.plot(months,sales_2024, linestyle = ':', color = 'blue', linewidth = 2, marker = '*', label = 'sales 2024')
plt.xlabel("Months")
plt.ylabel("Sales in 2024")
plt.title("Monthly sales data in FY 2024")
plt.legend(loc = 'upper left')
plt.grid(color = 'gray', linestyle = "-", linewidth = 0.5)
```
