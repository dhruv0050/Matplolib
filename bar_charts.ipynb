```python
import matplotlib.pyplot as plt
```

```python
## Vertical Bar Charts
category = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
values = [100,200,400,1000,50,300,600]
```

```python
plt.bar(category, values, color='orange', label = 'Sales 2025', width=0.5)
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Vertical Bar Chart Example')
plt.legend()
```

```python
## Horizontal Bar Charts
plt.barh(category, values, color='blue', label = 'Sales 2025', height=0.5)
plt.xlabel('Values')
plt.ylabel('Category')
plt.title('Horizontal Bar Chart Example')
plt.legend()
```
