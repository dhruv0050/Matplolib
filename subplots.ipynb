```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
## Method 1: Using plt.subplot(nrows,ncols,index)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, marker='s' ,color='blue', label='Line Chart')
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(1,2,2)
plt.bar(x,y, color='orange', label='Bar Chart')
plt.title('Bar Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.tight_layout()
plt.savefig('subplot_example1.png', dpi=300, bbox_inches='tight')
```

```python
## Method 2: Using plt.subplots(nrows,ncols,figsize=(width,height))
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(x,y,marker='o', color='green', label='Line Chart')
ax[0].set_title('Line Chart')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
ax[0].legend()

ax[1].bar(x,y,color='red', label='Bar Chart')
ax[1].set_title('Bar Chart')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')
ax[1].legend()
fig.suptitle('Subplots Example', fontsize=16)
plt.tight_layout()

plt.savefig('subplots_example2.png', dpi=300, bbox_inches='tight')
```
