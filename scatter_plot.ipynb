```python
import matplotlib.pyplot as plt
```

```python
hours_studied = [1, 2, 3, 4, 5, 6, 7]
marks_scored = [33, 45, 52, 60, 68, 75, 83]
```

```python
plt.scatter(hours_studied, marks_scored, color='blue', marker='^', label='Marks Scored')
plt.title('Scatter Plot of Hours Studied vs Marks Scored')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.grid(linestyle='--')
plt.legend()
```

```python
## Multiple scatter plot in one
```

```python
hours_studiedA = [1, 2, 3, 4, 5, 6, 7]
marks_scoredA = [33, 45, 52, 60, 68, 75, 83]
```

```python
hours_studiedB = [1, 2, 3, 4, 5, 6, 7]
marks_scoredB = [13, 30, 45, 55, 68, 70, 90]
```

```python
plt.scatter(hours_studiedA, marks_scoredA, color='purple', marker='*', label = "Marks scored by class A")
plt.scatter(hours_studiedB, marks_scoredB, color='coral', marker='^', label = "Marks scored by class B")
plt.title('Scatter Plot of Hours Studied vs Marks Scored by Class A and B')
plt.grid(linestyle='--')
plt.legend()
```
